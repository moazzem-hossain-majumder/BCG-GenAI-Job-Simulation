```python
import pandas as pd

df = pd.read_excel("BCG_GenAI_Financial_Data_Extraction_Example.xlsx", sheet_name="Extracted Data")
```

```python
df
```

```python
df = df.sort_values(['Company Name', 'Fiscal Year'])
df
```

```python
df['Revenue Growth (%)'] = (df.groupby('Company Name')['Total Revenue'].pct_change() * 100)
df['Revenue Growth (%)']
```

```python
df['Net Income Growth (%)'] = (df.groupby('Company Name')['Net Income'].pct_change() * 100)

df['Net Income Growth (%)']
```

```python
summary = df.groupby('Company Name').agg({'Revenue Growth (%)': 'mean', 'Net Income Growth (%)': 'mean'}).round(2)
print("\nAverage Growth Rates (%)")
print(summary)
```
